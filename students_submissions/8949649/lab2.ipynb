```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

**Matplotlib** : *Plotting Tangent function*


```python
X = np.linspace(0, 2*np.pi, 1000)
Y = np.tan(X)

fig, ax = plt.subplots()
ax.plot(X,Y, color = 'green')
```

**Matplotlib** : *Generate random data in 2D for the Subscribers company gained with time*

```python
# Generate random data for the x-axis (time) and y-axis (Subscribers)
time = list(range(20))
subscribers = np.random.randint(10,30, 20)

# Create the line graph
plt.plot(time, subscribers, marker='o')

# Add labels and title
plt.xlabel('Time (Years)')
plt.ylabel('Subscribers (Lakhs)')
plt.title('Subscribers Count')

# Display the graph
plt.show()
```

**Plotly** : *Combined statistical representations with px.histogram*

```python
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", color="sex",
                   marginal="box", # or violin, rug
                   hover_data=df.columns)
fig.show()
```
